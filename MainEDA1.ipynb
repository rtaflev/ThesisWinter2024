## Handle Missing Values

```python
# check for the cardinality to see if there are any categorical variables
df.nunique()
```

```python
df.columns
```

```python
# drop the columns with high missing values

df.drop(['Hepatitis B', 'GDP', 'Population'], axis=1, inplace=True)
```

```python
# Replace 0 with Nan so they are imputed further
df[['percentage expenditure','Income composition of resources','Schooling']].replace(0, np.nan, inplace=True)
```

- The Good tool is to fill the missing value with the mean of that column

```python
from sklearn.impute import SimpleImputer

# Select numerical columns
num_cols = df.select_dtypes(exclude='object').columns

# Apply SimpleImputer for numerical columns
imp = SimpleImputer(strategy='mean')
df[num_cols] = imp.fit_transform(df[num_cols])
```

```python
df.isnull().sum()
```

- Missing values are handled

```python
# change the status to developed for canada and france
df.loc[df['Country'] == 'Canada', 'Status'] = 'Developed'
df.loc[df['Country'] == 'France', 'Status'] = 'Developed'
```

```python
# Time Series analysis for overall Life Expectance
import seaborn as sns
sns.set_style("darkgrid")
plt.figure(figsize=(10,8))
sns.lineplot(x='Year',y='Life expectancy ',data=df,hue='Status')
plt.title('Life Expectancy over the years')
plt.ylabel('Life Expectancy')
plt.xlabel('Year')
plt.show()  
```

- Life Expectancy Depends heavily on Adult Mortality as well as BMI and HIV/Aids

```python
# plot the regplot between bmi and the life expectancy
plt.figure(figsize=(10,8))
sns.regplot(x='Life expectancy ',y='Income composition of resources',data=df)
plt.title('Life Expectancy vs Income composition of resources')
plt.show()
```
